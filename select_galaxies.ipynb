```python
import numpy as np
import astropy.io.fits as pyfits ## because I'm used to it!
import matplotlib.pyplot as plt
```

```python
## Change addresses to where you have downloaded the catalogs

cat_dir = '../WFIRST_WPS/CANDELS_fits/'
gs = pyfits.getdata(cat_dir+'gds.fits')
```

```python
gs.columns
```

```python
## Choose which galaxies you are interested in, wisely! 
sel = (gs['Spec_z']<1.5) &(gs['Spec_z']>0.2)&(gs['M_13a_tau']>10.0)&(gs['FWHM_IMAGE']>10)&(gs['FWHM_IMAGE']>15)&((gs['A_IMAGE_1']/gs['B_IMAGE_1'])<1.5)
```

```python
print (len(gs[sel]))
```

```python
ra,dec,redshift = gs['RA_1'][sel],gs['DEC_1'][sel],gs['Spec_z'][sel]

f=open('final_sample.cat','w+')
for boz in range(len(gs[sel])):
    f.write(np.str(ra[boz])+'\t'+np.str(dec[boz])+'\t'+np.str(redshift[boz])+'\n')
f.close()
```

```python
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation
from IPython.display import HTML
import matplotlib.colors as colors
import matplotlib.cm as cm

%matplotlib inline
```

```python
x, y = np.random.rand(2, 100) 
hist, xedges, yedges = np.histogram2d(x, y, bins=20, range=[[0, 20], [0, 20]])
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0
dx = dy = 0.5 * np.ones_like(zpos)

dz = []
for boz in range(100):
    h = 10+np.random.rand(400)
    hist = np.reshape(h,(20,20))
    dz.append(hist.ravel())

dz=np.array(dz)    
```

```python
fig, ax = plt.subplots()


def animate(i):
    ax = fig.add_subplot(111, projection='3d')
    ax.bar3d(xpos, ypos, zpos, dx, dy, dz[i,:], color='#cc6699', zsort='average')
    ax.view_init(elev=50., azim=210)
    ax.axis('off')


ani = matplotlib.animation.FuncAnimation(fig, animate, frames=10)#len(dz[:,0]))
HTML(ani.to_jshtml())
```
